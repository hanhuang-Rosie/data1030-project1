```python
def corr_heatmap(df, cols, title, filename = None):
    cols = [c for c in cols if c in df.columns]            
    Xsub = df[cols].copy()
    Xsub = Xsub.fillna(Xsub.median())                      
    C = Xsub.corr().values

    mask = np.triu(np.ones_like(C, dtype=bool), k=1)
    C_masked = np.ma.array(C, mask=mask)

    fig, ax = plt.subplots(figsize=(0.5+0.45*len(cols), 0.5+0.45*len(cols)))
    im = ax.imshow(C_masked, vmin=-1, vmax=1, aspect="auto")
    ax.set_xticks(range(len(cols)))
    ax.set_yticks(range(len(cols)))
    ax.set_xticklabels(cols, rotation=90)
    ax.set_yticklabels(cols)
    fig.colorbar(im, ax=ax, label="Pearson r")
    ax.set_title(title)
    plt.tight_layout()
    if filename is not None:
        fig.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()
corr_heatmap(X, socio_cols,  "Socioeconomic & Demographic (X-X)", filename="corr_socio.png")
corr_heatmap(X, police_cols, "Policing & Public Safety (X-X)",filename="corr_police.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def facet_hist_by_region(df, col, regions=("Northeast","Midwest","South","West"),
                         bins=30, to_percent=True, filename=None):
    
    x_all = df[col].astype(float)
    # auto-detect: if values look like 01 proportions, convert to %
    scale = 100.0 if (to_percent and np.nanmax(x_all.values) <= 1.0 + 1e-9) else 1.0
    label = f"{col} (%)" if scale == 100.0 else col

    fig, axes = plt.subplots(2, 2, figsize=(10, 6), sharex=True, sharey=True)
    axes = axes.ravel()

    # compute a common x-range across regions for fair comparison
    vals = x_all.to_numpy()
    lo = np.nanpercentile(vals, 0.5)
    hi = np.nanpercentile(vals, 99.5)
    # if all values identical (edge case)
    if not np.isfinite(lo) or not np.isfinite(hi) or lo == hi:
        lo, hi = np.nanmin(vals), np.nanmax(vals)

    for ax, r in zip(axes, regions):
        xr = df.loc[df["Region"]==r, col].astype(float).to_numpy()
        xr = xr[~np.isnan(xr)] * scale
        ax.hist(xr, bins=bins, range=(lo*scale, hi*scale), edgecolor="white")
        ax.set_title(f"{r} (n={xr.size})")
        ax.set_xlabel(label)
        ax.set_ylabel("count")

    fig.suptitle(f"{col} by Region", y=1.02, fontsize=12)
    plt.tight_layout()

    if filename is not None:
        fig.savefig(filename, dpi=300, bbox_inches="tight")
    
    plt.show()


facet_hist_by_region(X, "pctWhite", filename="hist_pctWhite_by_region.png")
facet_hist_by_region(X, "pctBlack", filename="hist_pctBlack_by_region.png")
facet_hist_by_region(X, "pctHisp", filename="hist_pctHisp_by_region.png")
```

```python
def xgb_model(X_train, Y_train, X_CV, y_CV, X_test, y_test, verbose=1):

    # make into row vectors to avoid an sklearn/xgb warning
    Y_train = np.reshape(np.array(Y_train), (1, -1)).ravel()
    y_CV = np.reshape(np.array(y_CV), (1, -1)).ravel()
    y_test = np.reshape(np.array(y_test), (1, -1)).ravel()

    XGB = xgb.XGBRegressor(n_jobs=1)
    
    # find the best parameter set
    param_grid = {"learning_rate": [0.03],
                  "n_estimators": [10000],
                  "reg_alpha": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
                  "reg_lambda": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
                  "missing": [np.nan], 
                  "max_depth": [1,3,10,30],
                  "colsample_bytree": [0.9],              
                  "subsample": [0.66]}

    pg = ParameterGrid(param_grid)

    scores = np.zeros(len(pg))

    for i in range(len(pg)):
        if verbose >= 5:
            print("Param set " + str(i + 1) + " / " + str(len(pg)))
        params = pg[i]
        XGB.set_params(**params,early_stopping_rounds=50)
        eval_set = [(X_CV, y_CV)]
        XGB.fit(X_train, Y_train,
                eval_set=eval_set, verbose=False)# with early stopping
        y_CV_pred = XGB.predict(X_CV)
        scores[i] = mean_squared_error(y_CV,y_CV_pred)

    best_params = np.array(pg)[scores == np.min(scores)]
    if verbose >= 4:
        print('Test set max score and best parameters are:')
        print(np.max(scores))
        print(best_params)

    # test the model on the test set with best parameter set
    XGB.set_params(**best_params[0],early_stopping_rounds=50)
    XGB.fit(X_train,Y_train,
            eval_set=eval_set, verbose=False)
    y_test_pred = XGB.predict(X_test)

    if verbose >= 1:
        print ('The MSE is:',mean_squared_error(y_test,y_test_pred))
    if verbose >= 2:
        print ('The predictions are:')
        print (y_test_pred)
    if verbose >= 3:
        print("Feature importances:")
        print(XGB.feature_importances_)

    return (mean_squared_error(y_test,y_test_pred), y_test_pred, XGB.feature_importances_)
```

```python
def knn_model(X_train, y_train, X_CV, y_CV, X_test, y_test, verbose=1):
    y_train = np.reshape(np.array(y_train), (1, -1)).ravel()
    y_CV = np.reshape(np.array(y_CV), (1, -1)).ravel()
    y_test = np.reshape(np.array(y_test), (1, -1)).ravel()

    model = Pipeline([
        
        ("knn", KNeighborsRegressor())
    ])

    param_grid = {
        "knn__n_neighbors": [3, 5, 11, 21],
        "knn__weights": ["uniform", "distance"],
        "knn__p": [1, 2]
    }

    pg = list(ParameterGrid(param_grid))
    scores = np.zeros(len(pg))

    for i, params in enumerate(pg):
        model.set_params(**params)
        model.fit(X_train, y_train)
        y_CV_pred = model.predict(X_CV)
        scores[i] = mean_squared_error(y_CV, y_CV_pred)

    best_params = pg[int(np.argmin(scores))]

    model.set_params(**best_params)
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    test_mse = mean_squared_error(y_test, y_test_pred)

    if verbose >= 1:
        print("Test MSE:", test_mse)
    if verbose >= 2:
        print("Best CV MSE:", float(np.min(scores)))
        print("Best params:", best_params)

    return test_mse, y_test_pred, None, best_params
```

```python
def ridge_model(X_train, y_train, X_CV, y_CV, X_test, y_test, verbose=1):
    y_train = np.asarray(y_train).ravel()
    y_CV    = np.asarray(y_CV).ravel()
    y_test  = np.asarray(y_test).ravel()

    model = Pipeline([
        ("ridge", Ridge())
    ])

    param_grid = {
        "ridge__alpha": [1e-3, 1e-2, 1e-1, 1, 10, 100],
        "ridge__fit_intercept": [True, False],
        "ridge__random_state": [0]
    }

    pg = list(ParameterGrid(param_grid))
    scores = np.zeros(len(pg))

    for i, params in enumerate(pg):
        model.set_params(**params)
        model.fit(X_train, y_train)
        y_CV_pred = model.predict(X_CV)
        scores[i] = mean_squared_error(y_CV, y_CV_pred)

    best_params = pg[int(np.argmin(scores))]

    model.set_params(**best_params)
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    test_mse = mean_squared_error(y_test, y_test_pred)

    coefs = model.named_steps["ridge"].coef_

    if verbose >= 1:
        print("Test MSE:", test_mse)
    if verbose >= 2:
        print("Best CV MSE:", float(np.min(scores)))
        print("Best params:", best_params)

    return test_mse, y_test_pred, coefs, best_params
```

```python
def decision_tree_model(X_train, y_train, X_CV, y_CV, X_test, y_test, verbose=1):
    y_train = np.asarray(y_train).ravel()
    y_CV    = np.asarray(y_CV).ravel()
    y_test  = np.asarray(y_test).ravel()

    model = DecisionTreeRegressor(random_state=0)

    param_grid = {
        "max_depth": [None, 3, 5, 10, 20],
        "min_samples_split": [2, 5, 10, 20],
        "min_samples_leaf": [1, 2, 5, 10]
    }

    pg = list(ParameterGrid(param_grid))
    scores = np.zeros(len(pg))

    for i, params in enumerate(pg):
        model.set_params(**params)
        model.fit(X_train, y_train)
        y_CV_pred = model.predict(X_CV)
        scores[i] = mean_squared_error(y_CV, y_CV_pred)

    best_params = pg[int(np.argmin(scores))]

    model.set_params(**best_params)
    model.fit(X_train, y_train)
    y_test_pred = model.predict(X_test)
    test_mse = mean_squared_error(y_test, y_test_pred)

    importances = model.feature_importances_

    if verbose >= 1:
        print("Test MSE:", test_mse)
    if verbose >= 2:
        print("Best CV MSE:", float(np.min(scores)))
        print("Best params:", best_params)

    return test_mse, y_test_pred, importances, best_params
```

```python
def reduced_feature_random_forest(X_train, y_train, X_CV, y_CV, X_test, y_test, verbose=1):
    # 1) unique missing patterns in test
    mask = X_test.isnull()
    unique_rows = np.array(np.unique(mask.values, axis=0))

    all_y_test_pred = pd.DataFrame()
    best_params_by_pattern = []

    if verbose:
        print('there are', len(unique_rows), 'unique missing value patterns.')

    # 2) loop patterns
    for i in range(len(unique_rows)):
        if verbose:
            print('working on unique pattern', i)

        # --- collect test rows that match this pattern ---
        sub_X_test = pd.DataFrame()
        sub_y_test = pd.Series(dtype=float)

        for j in range(len(mask)):
            if np.array_equal(np.array(mask.iloc[j]), unique_rows[i]):
                sub_X_test = pd.concat([sub_X_test, X_test.iloc[[j]]])
                sub_y_test = pd.concat([sub_y_test, y_test.iloc[[j]]])

        # keep only columns that are NOT missing for this pattern
        sub_X_test = sub_X_test[X_test.columns[~unique_rows[i]]]

        # --- reduced-feature train/CV ---
        sub_X_train = X_train[X_train.columns[~unique_rows[i]]]
        sub_X_CV    = X_CV[X_CV.columns[~unique_rows[i]]]

        # drop NaN rows
        sub_X_train = sub_X_train.dropna()
        sub_X_CV    = sub_X_CV.dropna()

        # align y (match your xgb style)
        sub_y_train = y_train.iloc[sub_X_train.index]
        sub_y_CV    = y_CV.iloc[sub_X_CV.index]

        # --- run RF via your random_forest_model ---
        # expected: test_mse, y_test_pred, importances, best_params
        sub_test_mse, sub_y_test_pred_arr, _, best_params = random_forest_model(
            sub_X_train, sub_y_train,
            sub_X_CV, sub_y_CV,
            sub_X_test, sub_y_test,
            verbose=0
        )

        best_params_by_pattern.append({
            "pattern_id": i,
            "n_test_rows": len(sub_X_test),
            "n_train_rows": len(sub_X_train),
            "n_cv_rows": len(sub_X_CV),
            "best_params": best_params
        })

        # store predictions with correct index
        sub_y_test_pred = pd.DataFrame(
            sub_y_test_pred_arr,
            columns=['sub_y_test_pred'],
            index=sub_y_test.index
        )

        if verbose:
            print('  RMSE:', np.sqrt(mean_squared_error(sub_y_test, sub_y_test_pred)))

        all_y_test_pred = pd.concat([all_y_test_pred, sub_y_test_pred])

    # 3) sort back to original order
    all_y_test_pred = all_y_test_pred.sort_index()
    y_test_sorted = y_test.sort_index()

    # 4) global metrics
    total_RMSE = np.sqrt(mean_squared_error(y_test_sorted, all_y_test_pred))
    total_R2   = r2_score(y_test_sorted, all_y_test_pred)

    return total_RMSE, total_R2, best_params_by_pattern, all_y_test_pred
```

```python
import numpy as np
import pandas as pd

def show_missing_patterns(df, name="DF"):
    
    mask = df.isna()
    
    
    patterns, inverse_idx = np.unique(mask.values, axis=0, return_inverse=True)
    counts = np.bincount(inverse_idx)

    print(f"\n==== Missing patterns in {name} ====")
    print(f"Total rows: {len(df)}, unique patterns: {len(patterns)}\n")

    for i, pattern in enumerate(patterns):
        cols_missing = df.columns[pattern]         
        n_rows = counts[i]
        print(f"Pattern {i+1}/{len(patterns)}")
        print(f"  rows: {n_rows}")
        print(f"  missing columns: {list(cols_missing) if len(cols_missing) > 0 else 'None'}\n")

        
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb


param_grid = {
    "learning_rate": [0.03],
    "n_estimators": [10000],
    "reg_alpha": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
    "reg_lambda": [0e0, 1e-2, 1e-1, 1e0, 1e1, 1e2],
    "missing": [np.nan], 
    "max_depth": [1,3,10,30,50],
    "colsample_bytree": [0.9],              
    "subsample": [0.66]
    }
pg = list(ParameterGrid(param_grid))
print(f"\nTotal param sets: {len(pg)}")

best_cv_mse = np.inf
best_params = None
best_model = None

for i, params in enumerate(pg, start=1):
    print(f"\nParam set {i}/{len(pg)}: {params}")

    model = xgb.XGBRegressor(
        objective="reg:squarederror",
        n_jobs=1,
        early_stopping_rounds=50,
        **params
    )

    model.fit(
        df_train, y_train,
        eval_set=[(df_CV, y_CV)],
        verbose=False
    )

    y_CV_pred = model.predict(df_CV)  
    cv_mse = mean_squared_error(y_CV, y_CV_pred)
    print("  CV RMSE:", np.sqrt(cv_mse))

    if cv_mse < best_cv_mse:
        best_cv_mse = cv_mse
        best_params = params
        best_model = model

print("\nBest CV RMSE:", np.sqrt(best_cv_mse))
print("Best params:", best_params)


y_test_pred = best_model.predict(df_test)
test_mse = mean_squared_error(y_test, y_test_pred)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_test_pred)

print("\nTest RMSE:", test_rmse)
print("Test R2:  ", test_r2)

feat_imp = pd.Series(best_model.feature_importances_, index=df_train.columns).sort_values(ascending=False)
print("\nTop 10 feature importances:")
print(feat_imp.head(10))
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import matplotlib.pyplot as plt

def xgb_importance_errorbars(df_train, y_train, df_CV, y_CV, best_params, K=10):
    feature_names = df_train.columns
    imps = []

    for s in range(K):
        model = xgb.XGBRegressor(
            objective="reg:squarederror",
            n_jobs=1,
            early_stopping_rounds=50,
            random_state=s,      
            **best_params
        )
        model.fit(df_train, y_train, eval_set=[(df_CV, y_CV)], verbose=False)
        imps.append(model.feature_importances_)

    imps = np.vstack(imps)  # shape: (K, p)
    mean_imp = imps.mean(axis=0)
    std_imp  = imps.std(axis=0, ddof=1)  # sample std

    imp_df = pd.DataFrame({
        "feature": feature_names,
        "mean": mean_imp,
        "std": std_imp
    }).sort_values("mean", ascending=False)

    top10 = imp_df.head(10).sort_values("mean", ascending=True)  

    # plot
    plt.figure(figsize=(10, 6))
    plt.barh(top10["feature"], top10["mean"], xerr=top10["std"])
    plt.xlabel("Feature importance (mean  std over seeds)")
    plt.title(f"Top 10 Feature Importances (XGB), K={K} seeds")
    plt.tight_layout()
    plt.savefig("xgb_top10_importance_errorbars.png", dpi=300, bbox_inches="tight")
    plt.show()

    return imp_df, imps


imp_df, imps = xgb_importance_errorbars(df_train, y_train, df_CV, y_CV, best_params, K=10)
```

```python
def plot_local_shap_bar(best_model, X, sample_idx=66, top_k=10):
    

    explainer = shap.TreeExplainer(best_model)
    shap_values = explainer.shap_values(X)  # (n, p)

    sv = shap_values[sample_idx, :]
    feature_names = np.array(X.columns)

    top_idx = np.argsort(np.abs(sv))[-top_k:]
    top_idx = top_idx[np.argsort(sv[top_idx])]  

    top_features = feature_names[top_idx]
    top_sv = sv[top_idx]

    base_value = explainer.expected_value
    pred_value = base_value + sv.sum()

    plt.figure(figsize=(12, 6))

    colors = np.where(top_sv >= 0, "tab:blue", "tab:red")
    plt.barh(top_features, top_sv, color=colors)

    plt.axvline(0, color="black", linewidth=1)
    plt.xlabel("SHAP value (impact on prediction)")
    plt.ylabel("Feature")

    plt.title(
        f"Local SHAP Explanation for Sample #{sample_idx}\n"
        f"Base value = {base_value:.1f}, Prediction = {pred_value:.1f}"
    )

    for y, v in enumerate(top_sv):
        if v >= 0:
            plt.text(v + 0.5, y, f"{v:.1f}", va="center")
        else:
            plt.text(v - 0.5, y, f"{v:.1f}", va="center", ha="right")

    plt.tight_layout()
    plt.savefig(f"local_shap_sample_{sample_idx}.png", dpi=300, bbox_inches="tight")
    plt.show()

plot_local_shap_bar(best_model, df_test, sample_idx=66, top_k=10)
plot_local_shap_bar(best_model, df_test, sample_idx=27, top_k=10)
```
